```python
from ast import literal_eval
```

```python
def extract_genres_and_keywords(obj):
    l = []
    for o in literal_eval(obj):
        l.append(o['name'])

    return l
```

```python
movies['genres'] = movies['genres'].apply(extract_genres_and_keywords)
movies['keywords'] = movies['keywords'].apply(extract_genres_and_keywords)
```

```python
movies
```

```python
def extract_first_three_cast_names(obj):
    return list(map(lambda cast: cast['name'], literal_eval(obj)[:3]))
```

```python
movies['cast'] = movies['cast'].apply(extract_first_three_cast_names)
```

```python
movies
```

```python
def extract_director_name(obj):
    filtered_director_crew = list(
        filter(lambda crew: crew['job'] == 'Director', literal_eval(obj)))
    return list(map(lambda item: item['name'], filtered_director_crew))


def replace_empty_space(obj):
    return [x.replace(" ", "") for x in obj]
```

```python
movies['crew'] = movies['crew'].apply(extract_director_name)
```

```python
movies['overview'] = movies['overview'].apply(
    lambda sentence: sentence.split())
```

```python
movies
```

```python
movies['genres'] = movies['genres'].apply(replace_empty_space)
movies['cast'] = movies['cast'].apply(replace_empty_space)
movies['cast'] = movies['cast'].apply(replace_empty_space)
movies['crew'] = movies['crew'].apply(replace_empty_space)
```

```python
movies
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + \
    movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movies = movies.drop(
    ['overview', 'genres', 'keywords', 'cast', 'crew'], axis=1)
```

```python
new_movies['tags'] = new_movies['tags'].apply(
    lambda item: " ".join(item)).apply(lambda item: item.lower())
```

```python
new_movies['tags']
```

```python
new_movies[new_movies['original_title'] == 'Batman']
```

```python
# Implement steming on similar words

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()


def stem_similar_strings(sentence):
    all_words = sentence.split()
    stem_words = list(map(lambda each_string: ps.stem(each_string), all_words))
    return " ".join(stem_words)
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem_similar_strings)
```

```python
new_movies['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')

movie_vector = cv.fit_transform(new_movies['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(movie_vector)
```

```python
similarity[2]
```

```python
new_movies[new_movies['original_title'] == 'Batman Begins']
```

```python
movie_index = new_movies[new_movies['original_title']== 'Batman'].index[0]
distances = similarity[movie_index]
```

```python
distances
```

```python
def recommend(movie_title):
    # get index of searched movie
    movie_index = new_movies[new_movies['original_title']== movie_title].index[0]
    # get distances of current movie from similarities
    distances = similarity[movie_index]
    recommendation = []
    for distance in sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]:
        index = distance[0]
        recommendation.append(new_movies.iloc[index]['original_title'])
    return recommendation
```

```python
recommend('Batman Begins')
```
